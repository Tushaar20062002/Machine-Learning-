```python
X = df.drop(columns=['price'],axis=1)
y = df['price']
```

```python
X
```

# 1)One Hot Encoding:

```python
X_one_encode = pd.get_dummies(X,columns=['model','transmission','fuelType'], drop_first=True, dtype=int)

X_one_encode
```

# 2) Labled encoding:

```python
from sklearn.preprocessing import LabelEncoder

X_labeled = X.copy()

columns = ['model', 'transmission', 'fuelType']

label_encoders = {}

for col in columns:
    le = LabelEncoder()
    X_labeled[col] = le.fit_transform(X_labeled[col].astype(str))
    label_encoders[col] = le

X_labeled.head()
```

# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerics_cols = ['year','mileage','tax','mpg','engineSize']
scaler = StandardScaler()

X_one_encode[numerics_cols] = scaler.fit_transform(X_one_encode[numerics_cols])

X_one_encode
```

```python
X_labeled.columns
```

```python
columns_to_scale = ['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize']

X_labeled[columns_to_scale] = scaler.fit_transform(X_labeled[columns_to_scale])
```

```python
X_labeled
```

## Model Training and Testing

1) for One hot Encoding

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_encode, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

2) for Labled Encoding

```python
X_train, X_test, y_train, y_test = train_test_split(X_labeled, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
r2 = r2_score(y_test, y_pred)
r2
```
